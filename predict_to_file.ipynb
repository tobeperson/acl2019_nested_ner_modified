```python
import json
def predict_to_file():
    """预测结果到文件，方便提交
    """
    fw = open("ner_pred_dev_v5.json", 'w', encoding='utf-8')
    with open("dev_system_predictions_v5.conll") as fr:
        event_list = []
        role=[]
        arguments=[]
        for l in fr.readlines(): 
            l="".join(l)
            l.rstrip("\n")
            if not l=="\n":
                argumentss=[]
                form, lemma, tag, label = l.split("\t")
                label= label.rstrip("\n")
                labels=label.split("|")
                for label in labels:
                    if label[0]=="B":
                        role.append(form)
                        arguments.append(label[2:])
                        continue
                    if label[0]=="I":
                        for key,argument in enumerate(arguments):
                            if  label[2:]==argument:
                                role[key]=role[key]+form
                        continue
                    if label[0]=="L":
                            for key,argument in enumerate(arguments):
                                print(role,arguments)
                                if  label[2:]==argument:
                                    role[key]=role[key]+form
                                    argumentss.append((argument,role[key]))
                                    del role[key]
                                    del arguments[key]
                    if label[0]=="U":
                        role1=form
                        argument1=label[2:]
                        argumentss.append((argument1,role1))
                    for (k, v) in argumentss:
                            event_list.append({
                                    'event_type': v,
                                    'arguments': [{
                                        'role': v,
                                        'argument': k
                                    }]
                                })
            else :
                l={}
                l['event_list'] = event_list
                print(l)
                l = json.dumps(l, ensure_ascii=False)
                event_list=[]
                role=[]
                arguments=[]
                fw.write(l + '\n')
        fw.close()
```
